```
import datetime
import pandas as pd
import FinanceDataReader as fdr
```

```
start = datetime.datetime(2019,1,1)
end = datetime.datetime(2019,6,30)

# Kakao(035720)
df_k = fdr.DataReader("035720", start, end)
df_k.head()
```

```
# Naver(035420)
df_n = fdr.DataReader("035420", start, end)
df_n.tail()
```

```
df_k['Volume'].plot(color='y') # Volume
```

```
df_n['Volume'].plot(color='g')
```

```
df_k = df_k[df_k.Volume > 0] # delete ROW when no transaction
df_k.head()
```

```
df_k['Close'].plot(figsize=(16,4), color='y', title = 'Kakao Price 2019 half') # price
```

```
df_k['Volume'].plot(figsize=(16,4), color='y',title= 'Kakao Volume 2019 half')
```

```
df_k2 = df_k[['Close', 'Volume']] # comparison
```

```
df_k2.head()
```

```
df_k2.plot()
```

```
df_k2['Close'].plot(figsize=(16,4), color='r')
df_k2['Volume'].plot(figsize=(16,4), color = 'b', secondary_y=True)
```

```
df_k2['2019-06':]['Close'].plot(color='r') # a specific point in time
df_k2['2019-06':]['Volume'].plot(color='b', secondary_y=True)
```

## Nomalization

```
df_k2.mean()
```

```
df_k_norm = (df_k2 - df_k2.mean()) / (df_k2.max() - df_k2.min())
df_k_norm.head()
```

```
df_k_norm.plot()
```
